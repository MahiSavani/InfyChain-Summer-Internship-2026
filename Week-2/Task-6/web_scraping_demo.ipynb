```python
import requests
from bs4 import BeautifulSoup
import time

session = requests.Session()
session.headers.update({
    "User-Agent"      : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                        "AppleWebKit/537.36 (KHTML, like Gecko) "
                        "Chrome/120.0.0.0 Safari/537.36",
    "Accept"          : "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language" : "en-US,en;q=0.5",
    "Referer"         : "https://www.google.com",  # pretend coming from Google
})

time.sleep(2)   # wait before requesting
response = session.get("https://www.ambitionbox.com/list-of-companies?page=1")
print("Status:", response.status_code)
print("Blocked?", "Access Denied" in response.text)
```

```python
soup      = BeautifulSoup(response.text, 'html.parser')
companies = soup.find_all('h2', class_='companyCardWrapper__companyName')
for company in companies:
    print(company.text.strip())
```

```python
ratings = soup.find_all('div', class_='rating_text rating_text--md')
for rating in ratings:
    print(rating.text.strip())
```

```python
reviews = soup.find_all('span', class_='companyCardWrapper__companyRatingCount')
for review in reviews:  
    print(review.text.strip())
```

```python
# loop through all cards on page 1
for card in soup.find_all('div', class_='companyCardWrapper'):
    
    inter = card.find('span', class_='companyCardWrapper__interLinking')
    
    if inter:
        parts    = inter.text.strip().split('|')
        industry = parts[0].strip() if len(parts) > 0 else "N/A"
        location = parts[1].strip() if len(parts) > 1 else "N/A"
    else:
        industry = "N/A"
        location = "N/A"
    
    print(f"Industry: {industry} | Location: {location}")
```

```python
for card in soup.find_all('div', class_='companyCardWrapper'):
    ratings  = card.find_all('span', class_='companyCardWrapper__ratingValues')

    print(f"Highly Rated: {ratings[0].text.strip() if len(ratings) > 0 else 'N/A'}")
    print(f"Critical: {ratings[1].text.strip() if len(ratings) > 1 else 'N/A'}")
```

```python
data = []
CompanyName = []
Rating = []
Reviews = []
Industry = []
Location = []
HighlyRated = []
CriticallyRated = []

for card in soup.find_all('div', class_='companyCardWrapper'):
    # Company Name
    company_name = card.find('h2', class_='companyCardWrapper__companyName')
    CompanyName.append(company_name.text.strip() if company_name else "N/A")
    
    Rating.append(card.find('div', class_='rating_text rating_text--md').text.strip() if card.find('div', class_='rating_text rating_text--md') else "N/A")
    # Ratings
    ratings = card.find_all('span', class_='companyCardWrapper__ratingValues')
    HighlyRated.append(ratings[0].text.strip() if len(ratings) > 0 else 'N/A')
    CriticallyRated.append(ratings[1].text.strip() if len(ratings) > 1 else 'N/A')
    
    # Reviews
    review_count = card.find('span', class_='companyCardWrapper__companyRatingCount')
    Reviews.append(review_count.text.strip() if review_count else "N/A")
    
    # Industry & Location
    inter = card.find('span', class_='companyCardWrapper__interLinking')
    if inter:
        parts = inter.text.strip().split('|')
        Industry.append(parts[0].strip() if len(parts) > 0 else "N/A")
        Location.append(parts[1].strip() if len(parts) > 1 else "N/A")
    else:
        Industry.append("N/A")
        Location.append("N/A")

data = {
    "CompanyName": CompanyName,
    "Reviews": Reviews,
    "Rating": Rating,
    "Industry": Industry,
    "Location": Location,
    "HighlyRated": HighlyRated,
    "CriticallyRated": CriticallyRated
}

```

```python
final = []   

for j in range(1,8 ):   

    print(f"Scraping page {j}...")

    session = requests.Session()
    session.headers.update({
        "User-Agent"      : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                            "AppleWebKit/537.36 (KHTML, like Gecko) "
                            "Chrome/120.0.0.0 Safari/537.36",
        "Accept"          : "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
        "Accept-Language" : "en-US,en;q=0.5",
        "Referer"         : "https://www.google.com",
    })

    response = session.get(
        f"https://www.ambitionbox.com/list-of-companies?page={j}"
    )

    if response.status_code != 200 or "Access Denied" in response.text:
        print(f"  ❌ Page {j} blocked — skipping")
        time.sleep(3)
        continue

    soup  = BeautifulSoup(response.text, 'lxml')
    cards = soup.find_all('div', class_='companyCardWrapper')

    print(f"  ✅ Found {len(cards)} companies")

    for card in cards:

        try:
            name = card.find('h2', class_='companyCardWrapper__companyName').text.strip()
        except:
            name = np.nan

        try:
            rating = card.find('span', class_='companyCardWrapper__companyRatingValue').text.strip()
        except:
            rating = np.nan

        try:
            reviews = card.find('span', class_='companyCardWrapper__companyRatingCount').text.strip()
        except:
            reviews = np.nan

        try:
            inter    = card.find('span', class_='companyCardWrapper__interLinking').text.strip()
            parts    = inter.split('|')
            industry = parts[0].strip()
            location = parts[1].strip() if len(parts) > 1 else np.nan
        except:
            industry = np.nan
            location = np.nan

        try:
            ratings  = card.find_all('span', class_='companyCardWrapper__ratingValues')
            highly   = ratings[0].text.strip()
        except:
            highly   = np.nan

        try:
            
            ratings  = card.find_all('span', class_='companyCardWrapper__ratingValues')
            critical = ratings[1].text.strip()
        except:
            critical = np.nan

        # ── Append each card as dict ───────────────────────
        final.append({
            "Company"          : name,
            "Rating"           : rating,
            "Reviews"          : reviews,
            "Industry"         : industry,
            "Location"         : location,
            "Highly_Rated"     : highly,
            "Critically_Rated" : critical,
            "Page"             : j        # track which page it came from
        })

    time.sleep(2)

df_final = pd.DataFrame(final)
print(f"\n✅ Total companies scraped: {len(df_final)}")
print(df_final.head(10))

df_final.to_csv("ambitionbox_all_companies.csv", index=False)
print("✅ Saved to ambitionbox_all_companies.csv")
```
